```python
import pandas as pd

path = "...FDNY\\"

df = pd.read_csv(path + "NYC_Fire_Department_Building_Vacate_List_20260416.csv", low_memory=False)

df.BBL = df.BBL.astype(str)

dp = pd.read_csv(path + "pluto_25v4.csv", low_memory=False)
```

```python
# Borough mapping
boro_map = {
    'MN': '1',
    'BX': '2',
    'BK': '3',
    'QN': '4',
    'SI': '5'
}

# Create BBL
dp['BBL'] = (
    dp['borough'].map(boro_map) +
    dp['block'].astype(int).astype(str).str.zfill(5) +
    dp['lot'].astype(int).astype(str).str.zfill(4)
)

dp.BBL = dp.BBL.astype(str)
```

Merge

```python
dp = dp[["BBL", "latitude", "longitude", "yearbuilt"]]

lat = dp.set_index('BBL')['latitude'].to_dict()
df["lat"] = df["BBL"].map(lat)

lon = dp.set_index('BBL')['longitude'].to_dict()
df["lon"] = df["BBL"].map(lon)

yb = dp.set_index('BBL')['yearbuilt'].to_dict()
df["Year Built"] = df["BBL"].map(yb)
```

```python
Create Building Age Category
```

```python
def age_category(year):
    if pd.isna(year):
        return 'Unknown'
    elif year >= 2000:
        return 'Post 2000'
    elif year >= 1980:
        return '1980 to 1999'
    elif year >= 1960:
        return '1960 to 1979'
    elif year >= 1940:
        return '1940 to 1959'
    elif year >= 1920:
        return '1920 to 1939 (Pre-war late)'
    elif year >= 1900:
        return '1900 to 1919 (Pre-war mid)'
    else:
        return 'Pre 1900'

df['Building Age Category'] = df['Year Built'].apply(age_category)
```
